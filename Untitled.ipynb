```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def crawling(soup):
    result = []
    for row in soup.find_all('tr', onmouseover=lambda x: x is not None):
        cols = row.find_all('td', class_='num')
        date = row.find('span', class_='tah p10 gray03').get_text().strip()
        values = [date] + [col.get_text().strip().replace(',', '') for col in cols]
        result.append(values)
    return result

company_code = '005930'  # 삼성전자
url = "https://finance.naver.com/item/sise_day.nhn?code=" + company_code

headers = {
    'referer': 'https://finance.naver.com/item/sise.naver?code=005930',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}
response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")

data = crawling(soup)

columns = ["날짜", "종가", "전일비", "시가", "고가", "저가", "거래량"]
df = pd.DataFrame(data, columns=columns)

print(df)
```
